```python
import pandas as pd
from sqlalchemy import create_engine, text
from IPython.display import display

# Kết nối tới PostgreSQL
engine = create_engine('postgresql+psycopg2://airflow:airflow@localhost:5434/airflow')

# Danh sách bảng
tables = [
    'brands', 'categories', 'loyalty_programs', 'suppliers', 'products',
    'stores', 'employees', 'payments', 'sales_transactions', 'sales_items', 'inventory',
    'stock_movements', 'purchase_orders', 'shipments', 'returns', 'promotions',
    'campaigns', 'customer_feedback', 'store_visits', 'pricing_history', 'discount_rules',
    'tax_rules'
]

# Schema nếu cần (ví dụ: raw_retails_data)
schema = 'raw_retails_data'

# Danh sách kết quả
results = []

customers_df = pd.read_sql(f'SELECT customer_id, name, email, phone, loyalty_program_id, created_at FROM snapshots.snapshot_customers', engine)
customers_num_rows = customers_df.shape[0]
customers_num_cols = customers_df.shape[1]
customers_nulls = customers_df.isnull().any(axis=1).sum()  

results.append({
            'Table': 'customers',
            'Num_cols': customers_num_cols,
            'Num_rows': customers_num_rows,
            'Nulls': customers_nulls
        })


for table in tables:
    try:
        # Đọc bảng
        df = pd.read_sql(f'SELECT * FROM {schema}.raw_{table}', engine)

        # Tính toán
        num_rows = df.shape[0]
        num_cols = df.shape[1]
        nulls = df.isnull().any(axis=1).sum()

        results.append({
            'Table': table,
            'Num_cols': num_cols,
            'Num_rows': num_rows,
            'Nulls': nulls
        })
    except Exception as e:
        results.append({
            'Table': table,
            'Num_cols': 'Error',
            'Num_rows': 'Error',
            'Nulls': str(e)
        })

# Chuyển thành DataFrame và hiển thị
summary_df = pd.DataFrame(results)
display(summary_df)

```
